```python
# importing the required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# importing the dataset

df = pd.read_csv('/content/Final Consolidated data set -ANAC2.csv',encoding='ISO-8859-1')
df.head()
```

```python
# shape of dataset
df.shape
```

```python
# checking null values
df.isnull().sum()
```

```python
df['Creator Gender'].unique()
```

```python
df['Creator Gender']
```

```python
# impute null values

df['Creator Gender'].fillna('Company',inplace=True)
```

```python
df['Creator Gender'].unique()
```

```python
df.dropna(axis=0,how='any',inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.reset_index(inplace=True)
```

```python
df.columns
```

```python
df.drop(['index','Duration of Video'], axis=1)
```

```python
df.info()
```

```python
pivot_language = df[df['Language of the Video'].isin(['English','Hindi','Malayalam','Tamil','Telugu'])].pivot_table(values='Video Views', index='Language of the Video', aggfunc='mean', sort=True)
pivot_language
```

```python
# Language analysis
pivot_language.plot(kind='bar')
plt.title('Language wise analysis')
```

```python
pivot_videoquality = df.pivot_table(values='Video Views', index='Maximum Quality of the Video', aggfunc='mean', sort=True)
pivot_videoquality
```

```python
# Video quality analysis

pivot_videoquality.plot(kind='bar')
plt.title('Video quality analysis')
```

```python
df['binned']=pd.cut(df['Duration in Seconds'],bins=300)
```

```python
pivot_duration = df.pivot_table(values='Video Views', index='binned', aggfunc='mean', sort=True)

pivot_duration.plot()
plt.xticks(rotation=90)
plt.xlabel('Duration in seconds binned')
```
